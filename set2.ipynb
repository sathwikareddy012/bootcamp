```python
#merge 2 dictionaries
d1={'a':12,'b':34,'c':56}
d2={'b':24,'c':78,'d':112}
for key in d2:
    if key in d1:
        d1[key]=d1[key] + d2[key]
    else:
        d1[key] = d2[key]
        print(d1)
```

```python
#find key with maximum values
d={'a':10,'b':20,'c':30}
max_value = max(d.values())
for key in d:
    if d[key] == max_value:
        print(key)
```
