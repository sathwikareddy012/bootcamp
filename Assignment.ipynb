```python
students = []

def add_student():
    roll = int(input("Enter Roll Number: "))
    name = input("Enter Name: ")
    marks = int(input("Enter Marks: "))

    student = {
        "roll": roll,
        "name": name,
        "marks": marks
    }

    students.append(student)
    print("Student added successfully")


def view_students():
    for s in students:
        print(s)


def search_student():
    r = int(input("Enter Roll Number: "))

    for s in students:
        if s["roll"] == r:
            print(s)
            return

    print("Student not found")


def delete_student():
    r = int(input("Enter Roll Number to delete: "))

    for s in students:
        if s["roll"] == r:
            students.remove(s)
            print("Deleted successfully")
            return

    print("Student not found")


def calculate_average():
    total = 0

    for s in students:
        total += s["marks"]

    print("Class Average:", total / len(students))


def assign_grade():
    for s in students:
        m = s["marks"]

        if m >= 75:
            s["grade"] = "Distinction"
        elif m >= 60:
            s["grade"] = "First Class"
        elif m >= 50:
            s["grade"] = "Second Class"
        else:
            s["grade"] = "Fail"

    print("Grades assigned")


def topper():
    max_marks = 0

    for s in students:
        if s["marks"] > max_marks:
            max_marks = s["marks"]

    for s in students:
        if s["marks"] == max_marks:
            print("Topper:", s)


while True:

    print("\n===== Student Performance Management System =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Delete Student")
    print("5. Calculate Average")
    print("6. Assign Grades")
    print("7. Find Topper")
    print("8. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        add_student()

    elif choice == 2:
        view_students()

    elif choice == 3:
        search_student()

    elif choice == 4:
        delete_student()

    elif choice == 5:
        calculate_average()

    elif choice == 6:
        assign_grade()

    elif choice == 7:
        topper()

    elif choice == 8:
        print("Exit")
        break

    else:
        print("Invalid choice")
```
