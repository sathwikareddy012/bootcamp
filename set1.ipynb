```python
x=[45,75,27,58,60,80,76,46,87,86]
print(x)
```

```python
#find MAXIMUM
max=x[0]
for i in range(1,len(x),1):
    if x[i] > max:
        max=x[i]
        id=i
        print(max,i)
    else:
        continue
print(f"the maximum is {max} at index {id}")

```

```python
max = x[0]
for j in x:
    if j > max:
        max=j
print(f"the maimum is {max}")
```

```python
#find MINIMUM
min=x[0]
for i in range(1,len(x),1):
    if x[i] < min:
        min=x[i]
        id=i
        print(min,i)
    else:
        continue
print(f"the minimum is {min} at index {id}")
```
